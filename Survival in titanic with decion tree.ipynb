```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked','PassengerId'],axis='columns',inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
```

```python
#create dummy for sex
df['Sex_n']=le_Sex.fit_transform(df['Sex'])
```

```python
df.drop(['Sex'],axis='columns',inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#fillna value
df.Age=df.Age.fillna(df.Age.mean())
```

```python
#convert float to int
df[['Age','Fare']]=df[['Age','Fare']].astype(int)
```

```python
df
```

```python
df.drop(['Survived'],axis='columns')
target=df['Survived']
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(df,target)
```

```python
model.score(df,target)
```

```python
model.predict([[3,35,13,1,0]])
```
